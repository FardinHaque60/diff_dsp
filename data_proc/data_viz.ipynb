## inspect data set


```python
import pickle
import numpy as np
import matplotlib.pyplot as plt
import random

with open("../data/train_set.pickle", "rb") as f:
    train_set = pickle.load(f)

with open("../data/test_set.pickle", "rb") as f:
    test_set = pickle.load(f)

for i in range(10):
    r = random.randint(1, 30)
    item = train_set[r]
    print(item.shape)

    train = item[0]
    test = item[1]

    plt.figure(figsize=(8, 4))
    plt.plot(train)
    plt.plot(test)
    plt.title("item in set")
    plt.xlabel("Index")
    plt.ylabel("Value")
    plt.show()
```
